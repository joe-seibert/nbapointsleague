```python
lg_pts = {}

for n in league:
    lg_pts.setdefault(n, [])
    for p in league[n]:
        try:
            pstats = player_stats[['PLAYER_NAME','PTS','GP']].query(f'PLAYER_NAME=="{p}"').to_numpy()[0]
            lg_pts[n].append(tuple(pstats))
        except IndexError:
            pstats = (p, 0, 0)
            lg_pts[n].append(pstats)
    lg_pts[n] = np.array(lg_pts[n],
                         dtype=[('player_name', 'U50'), ('points', 'i4'), ('games_played', 'i4')])
```
